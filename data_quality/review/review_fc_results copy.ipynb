```python
symbol = "#NVDA"
import pandas as pd
import matplotlib.pyplot as plt

# Beide Dateien einlesen (hier als Beispiel mit gleichem Dateinamen)
df_forecast = pd.read_csv("/Users/Shared/ai_work/Trainingdata/ml_results/lstm_ta/" + symbol + ".csv")
df_histories = pd.read_csv("/Users/Shared/ai_work/Trainingdata/ml_data/yh_his/D1/" + symbol + ".csv")
df_sk_linear = pd.read_csv("/Users/Shared/ai_work/Trainingdata/ml_results/sk_linear/" + symbol + ".csv")

# Datumsspalte in Datetime konvertieren
df_forecast['date'] = pd.to_datetime(df_forecast['Date'])
df_forecast['close'] = df_forecast['Close'].astype(float)

df_sk_linear['date'] = pd.to_datetime(df_sk_linear['Date'])
df_sk_linear['close'] = df_sk_linear['Close'].astype(float)

df_histories['date'] = pd.to_datetime(df_histories['date'])
df_histories = df_histories.tail(100)
plt.figure(figsize=(14, 7))

# Forecast (grn)
plt.plot(df_forecast['date'], df_forecast['close'], label='LSTM-TA', color='green')

# Histories (magenta)
plt.plot(df_histories['date'], df_histories['close'], label='Histories', color='magenta')

# SK Linear (blau)
plt.plot(df_sk_linear['date'], df_sk_linear['close'], label='SK Linear', color='blue')

plt.xlabel('Datum')
plt.ylabel('Close')
plt.title( symbol + ' Forecast vs. Histories')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
